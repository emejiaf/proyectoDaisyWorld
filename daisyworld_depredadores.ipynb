```python
class Planeta:
    
    def __init__(self, plantas, animales, albedo_f, luminosidad, prop_fertil, q = 15, boltz = 5.67e-8):
        self.blanca = plantas[0]
        self.negra = plantas[1]
        self.cheat = plantas[2]
        self.conejo = animales[0]
        self.zorro = animales[1]
        self.albedo_f = albedo_f
        self.luminosidad = luminosidad
        self.boltz = boltz
        self.albedo_p = None
        self.temp_g = []
        self.temp_g_vacio = []
        self.q = q
        self.prop_fertil = prop_fertil
        self.area_blanca = []
        self.area_negra = []
        self.area_cheat = []
        self.pob_conejo = []
        self.pob_zorro = []
        self.step = 0
        
    def actualiza_temp_g(self):
        self.temp_g = np.append(self.temp_g,((self.luminosidad[self.step]/(self.boltz))*(1 - self.albedo_p))**0.25 )
        
    def temp_planeta_vacio(self):
        for i in self.luminosidad:
            self.temp_g_vacio = np.append(self.temp_g_vacio,((i/(self.boltz))*(0.5))**0.25 )
    
    def actualiza_albedo_p(self):
        self.prop_fertil = 1 - self.blanca.prop_area - self.negra.prop_area - self.cheat.prop_area
        self.albedo_p = self.blanca.prop_area*self.blanca.albedo + self.negra.prop_area*self.negra.albedo + (self.prop_fertil + self.cheat.prop_area)*self.albedo_f

    def vivir(self):
        for i in self.luminosidad:
            self.actualiza_albedo_p()
            self.actualiza_temp_g()
            self.blanca.actualizar(self)
            self.negra.actualizar(self)
            self.cheat.actualizar(self)
            self.conejo.actualiza_conejo(self)
            self.zorro.actualiza_zorro(self)
            self.area_blanca = np.append(self.area_blanca,blanca.prop_area)
            self.area_negra = np.append(self.area_negra,negra.prop_area)
            self.area_cheat = np.append(self.area_cheat,cheat.prop_area)
            self.pob_conejo = np.append(self.pob_conejo, conejo.poblacion)
            self.pob_zorro = np.append(self.pob_zorro, zorro.poblacion)
            self.step += 1
    
    def graficar_mundo(self):
        grafica_mundo, ax = plt.subplots(3, sharex = True,figsize=(14,20))
        ax[0].plot(self.luminosidad, self.area_blanca, linewidth = .5)
        ax[0].plot(self.luminosidad, self.area_negra, color = "black", linewidth = .5)
        ax[0].plot(self.luminosidad, self.area_cheat, color  = "grey", linewidth = .5)
        ax[0].plot(self.luminosidad, self.area_blanca + self.area_negra + self.area_cheat, color = "red", linewidth = 1.5)
        ax[0].set_ylabel("Proporcion de area cubierta")
        ax[0].legend(["Area blanca","Area negra","Area gris","Area total"])
        ax[1].plot(self.luminosidad,self.pob_conejo, color = "blue", linewidth = 1.5, label="Conejos")
        ax[1].plot(self.luminosidad,self.pob_zorro, color = "red", linewidth = 1.5, label = "Zorros")
        ax[1].legend(["Conejos","Zorros"])
        ax[1].set_ylabel("Poblacion")
        ax[2].plot(self.luminosidad,self.temp_g, color = "blue", linewidth = 1.5, label = "DaisyWorld")
        ax[2].plot(self.luminosidad, self.temp_g_vacio, color = "red", linewidth = 1.5, label = "Mundo vacio")
        ax[2].legend(["DaisyWorld","Mundo Vacio"])
        ax[2].set_ylabel("Temperatura global (K)")
        ax[2].set_xlabel("Luminosidad solar")
        
    def graficar_temp_g(self):
        plt.figure(figsize=(8, 6), dpi=80)
        plt.plot(self.luminosidad,self.temp_g, color = "blue", linewidth = 1.5, label = "DaisyWorld")
        plt.plot(self.luminosidad, self.temp_g_vacio, color = "red", linewidth = 1.5, label = "Mundo vacio")
        plt.legend(loc='upper left')
        plt.ylabel("Temperatura global (K)")
        plt.xlabel("Luminosidad solar") 
```

```python
class Planta:
    dif_area =0
    temp_l=0
    tasa_cre = 0
    def __init__(self, color, albedo, prop_area, tasa_muerte = 0.1, superior = 313, inferior = 278):
        self.color = color
        self.albedo = albedo
        self.prop_area = prop_area
        self.tasa_muerte = tasa_muerte
        self.superior = superior
        self.inferior = inferior
           
    def actualiza_temp_l(self, mundo):
        self.temp_l = mundo.q * (mundo.albedo_p - self.albedo) + mundo.temp_g[mundo.step]
        
    def actualiza_taza_cre(self):
        if self.temp_l > self.inferior and self.temp_l < self.superior :
            self.tasa_cre = (self.temp_l - self.inferior) * (self.superior - self.temp_l) * 4 / (self.inferior -self.superior)**2
        else:
            self.tasa_cre = 0          
            
    def actualiza_area(self,mundo):
        self.dif_area = self.prop_area*(self.tasa_cre * mundo.prop_fertil - self.tasa_muerte)
        self.prop_area = self.prop_area  + self.dif_area
        if self.prop_area <= 0.01:
            self.prop_area = 0.01
        if self.prop_area >= .99:
            self.prop_area = .99

    def actualizar(self,mundo):
        self.actualiza_temp_l(mundo)
        self.actualiza_taza_cre()
        self.actualiza_area(mundo)
```

```python
class animal:
    def __init__(self, plantas, poblacion, poblacion_min, tasa_muerte, disponibilidad, tasa_cre, const_biomasa):
        self.blanca = plantas[0]
        self.negra = plantas[1]
        self.cheat = plantas[2]
        self.poblacion = poblacion
        self.poblacion_min = poblacion_min
        self.tasa_muerte = tasa_muerte
        self.disponibilidad = disponibilidad
        self.tasa_cre = tasa_cre
        self.const_biomasa = const_biomasa
    
        
    def actualiza_conejo(self, mundo): 
        self.dif = self.tasa_cre *self.poblacion*((self.disponibilidad*(mundo.blanca.prop_area + mundo.negra.prop_area + mundo.cheat.prop_area))**2 - self.tasa_muerte);
        self.poblacion = self.poblacion + self.const_biomasa*self.dif
        if self.dif >0:
            mundo.blanca.prop_area -= mundo.blanca.prop_area * self.dif
            mundo.negra.prop_area -= mundo.negra.prop_area * self.dif
            mundo.cheat.prop_area -= mundo.cheat.prop_area * self.dif        
        if self.poblacion < self.poblacion_min:
            self.poblacion = self.poblacion_min

                
    def actualiza_zorro(self, mundo): 
        self.dif = self.tasa_cre*self.poblacion*((self.disponibilidad*mundo.conejo.poblacion)**2 - self.tasa_muerte);
        self.poblacion = self.poblacion + self.const_biomasa*self.dif
        if self.dif >0:
            mundo.conejo.poblacion -= mundo.conejo.poblacion * self.dif
        if self.poblacion < self.poblacion_min:
            self.poblacion = self.poblacion_min

        
```

```python
area_n = np.random.rand()
area_b = np.random.rand()
area_g = np.random.rand()
while (area_n + area_b + area_g) > 1:
    area_n = np.random.rand()
    area_b = np.random.rand()
    area_g = np.random.rand()
area_f = 1 - (area_n + area_b + area_g)

negra = Planta("negra",.25, area_n)
cheat = Planta("cheat", 0.5, area_g)
blanca = Planta("blanca",.75, area_b)

luminosidad = np.linspace(.6, 1.8, 1000)
luminosidad *= 1000
plantas = [negra, blanca, cheat]

pob_z = np.random.rand()
pob_c = np.random.rand()

conejo = animal(plantas, poblacion = pob_c, poblacion_min = .005, tasa_muerte = .2, disponibilidad = .8, tasa_cre =  2.5, const_biomasa = .3)
zorro = animal(plantas, poblacion = pob_z, poblacion_min = .01, tasa_muerte = .4, disponibilidad = .4, tasa_cre = .8, const_biomasa = .4)

animales = [conejo, zorro]

mundo_cruel = Planeta(plantas, animales, 0.5, luminosidad, area_f)
mundo_cruel.vivir()
mundo_cruel.temp_planeta_vacio()
mundo_cruel.graficar_mundo()
```
